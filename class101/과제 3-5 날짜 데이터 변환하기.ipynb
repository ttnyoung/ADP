# 날짜 데이터 전처리 과제 
- 데이터 : "6.세종시_오피스텔(매매)_실거래가.csv"

```python
import pandas as pd 
df_sejong = pd.read_csv("./data/6.세종시_오피스텔(매매)_실거래가.csv",encoding='utf8')
```

```python
df_sejong
```

```python
### 아래와 같이 계약년월을 datetime으로 변경하세요. 
```

```python
from datetime import datetime

df_sejong.info()
```

```python
df_sejong['계약년월'].astype(str) + '_' + df_sejong['계약일'].astype(str)
```

```python
df_sejong['계약년월'] = pd.to_datetime(df_sejong['계약년월'].astype(str) + '_' + df_sejong['계약일'].astype(str), format = '%Y%m_%d')
df_sejong
```

```python
## 계약년월을 아래와 같이 문자열로 다시 변경하세요. 
```

```python
df_sejong['계약년월'] = df_sejong['계약년월'].dt.strftime('%Y_%m_%d')
df_sejong
```

```python
## 6월 데이터만 추출해보세요 
```

```python
df_sejong.info()
```

```python
df_sejong[df_sejong['계약년월'].map(lambda x: datetime.strptime(x, '%Y_%m_%d').month).isin([6])]
```

```python
## 날짜순으로 데이터를 정렬해보세요 
```

```python
df_sejong.sort_values(by='계약년월')
```
