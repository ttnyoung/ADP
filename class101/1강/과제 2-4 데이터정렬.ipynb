## players_20.csv데이터에서 height_cm, weight_kg 의 중위값(median)을 출력하세요.

```python
import pandas as pd 
```

```python
fifa_df = pd.read_csv("/content/drive/MyDrive/ADP24/class101/2.데이터_핸들링/data/players_20.csv")
#'/content/drive/MyDrive/ADP24/class101/2.데이터_핸들링/data/players_20.csv'
fifa_df.head()
```

```python
h_median = fifa_df['height_cm'].median()
w_median = fifa_df['weight_kg'].median()
h_median, w_median
```

## 데이터셋의 height_cm, weight_kg 이 중위값(median)보다 큰 선수를 출력하세요 

```python
fifa_df[(fifa_df['height_cm']>h_median) & (fifa_df['weight_kg']>w_median)]
```

## height_cm 기준으로 내림차순 정렬하세요

```python
fifa_df.sort_values(by='height_cm',ascending=False)
```

## weight_kg 기준으로 내림차순 정렬하세요

```python
fifa_df.sort_values(by='weight_kg',ascending=False)
```

## height_cm 와 weight_kg 를 기준으로 오름차순 정렬하세요 

```python
fifa_df.sort_values(by=['height_cm','weight_kg'],ascending=True)
```
