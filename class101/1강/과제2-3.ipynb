```python
import pandas as pd
import numpy as np
```

```python
filepath = '/content/drive/MyDrive/ADP24/class101/2.데이터_핸들링/data/7.세종시_아파트(전월세)_실거래가.csv'
df_apart = pd.read_csv(filepath)
df_apart.head()
```

```python
#1) 단지명 중  '해밀마을 2단지'인 row 삭제 
```

```python
df_apart1 = df_apart[df_apart['단지명']!='해밀마을 2단지']
df_apart1
```

```python
#2) new_시군구' 컬럼을 만들고 '세종특별자치시 고운동' => '세종특별자치시 해밀동' 으로 변경
```

```python
df_apart1['new_시군구'] = df_apart1['시군구']
```

```python
df_apart1.loc[df_apart1['new_시군구']=='세종특별자치시 고운동','new_시군구'] = '세종특별자치시 해밀동'
```

```python
df_apart1
```

```python
#3) 'Class'컬럼을 만든 뒤 전세면 True 아니면 False로 입력 (loc사용) 
```

```python
df_apart1.loc[df_apart['전월세구분']=="전세",'Class']= True
df_apart1.loc[df_apart['전월세구분']!="전세",'Class']= False
df_apart1
```

```python
#4) 전세이면서 보증금이 1억 이상인 데이터만을 추출 
```

```python
df_apart1[(df_apart1['Class']==True) & (df_apart1['보증금(만원)']>=10000)]
```

```python
#5) -10~0층 : 지하 , 1~5층 : 저층 6~15층 : 중층, 16층 이상 : 고층으로 할당하는 새로운 컬럼 생성 
```

```python
df_apart1.loc[df_apart1['층']<=0,'new_층'] = "지하"
df_apart1.loc[df_apart1['층']>=1,'new_층'] = "저층"
df_apart1.loc[df_apart1['층']>=6,'new_층'] = "중층"
df_apart1.loc[df_apart1['층']>=16,'new_층'] = "고층"
df_apart1
```
