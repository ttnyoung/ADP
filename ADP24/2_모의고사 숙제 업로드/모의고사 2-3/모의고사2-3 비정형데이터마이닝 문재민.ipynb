# Q1

```python
# 1) ‘연설문.txt’ 데이터를 읽어온 뒤 숫자, 특수 문자 등을 제거하는 전처리 작업을 시행하시오.
```

```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('./연설문.txt', sep='\n', engine='python', encoding='cp949', header=None)
df.columns = ['sentence']
```

```python
df = pd.DataFrame(df['sentence'].map(lambda x: re.sub("[^a-zA-Zㄱ-ㅎ가-힣\s]","",x)))
df
```

# Q2

```python
# 2) 전처리된 데이터에서 명사를 추출하고 명사의 출현빈도를 10위까지 추출하여 막대그래프로 시각화하시오.
```

```python
list(df.sentence)
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
print(okt.nouns(df['sentence'].sum()))
```

```python
nouns = okt.nouns(df['sentence'].sum())
nouns = [n for n in nouns if len(n)>1]
nouns
```

```python
df_nouns = pd.DataFrame(nouns,columns = ['noun'])
df_nouns = pd.DataFrame(df_nouns.value_counts()).reset_index()
df_nouns.columns = ['noun','counts']
df_nouns[:10]
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.bar(df_nouns[:10].noun, df_nouns[:10].counts)
```

# Q3

```python
# 3) 전처리된 데이터를 이용해 워드클라우드를 작성하고 인사이트를 추출하시오.
```

```python
from collections import Counter

words = Counter(nouns)
words.most_common(10)
```

```python
from wordcloud import WordCloud

wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250)

gen = wc.generate_from_frequencies(words)
plt.figure(figsize=(10,10))
plt.imshow(gen)
```
