```python
#---------------------------------------------------------------------------
# Q2) ‘영화 기생충_사전.txt’를 단어 사전으로 하는 TDM을 구축하고 빈도를 파악하고 시각화하라.
#---------------------------------------------------------------------------
```

```python
df_dic = pd.read_csv('./영화 기생충_사전.txt', sep='\n', engine='python', encoding='cp949', header=None)
df_dic
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
tdm = cv.fit_transform(df_1['review'].str.replace("[\.\,/\?\;\:\~\!\@\#\$\%\^\&\*]",""))
```

```python
tdm
```

```python
print(tdm)
```

```python
cv.get_feature_names()
```

```python
tdm.sum(axis=0)
```

```python
word_count = pd.DataFrame({'빈도':tdm.sum(axis=0).flat, '단어':cv.get_feature_names_out()})
word_count.sort_values(by='빈도', ascending=False, inplace=True)
a = word_count[word_count['단어'].isin(df_dic[0])]
a.reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
plt.bar(a['단어'],a['빈도'])
```
