# 과제 3-3 문자열 데이터 변환하기 (str) 

## 1 '6.세종시_오피스텔(매매)_실거래가.csv' 데이터 불러오기 


```python
import pandas as pd
import numpy as np
df_sejong = pd.read_csv('./data/6.세종시_오피스텔(매매)_실거래가.csv')
df_sejong
```

## 2. 시군구 띄어쓰기로 3개의 컬럼으로 변경 


```python
df_sejong_split = df_sejong['시군구'].str.split(' ',expand=True)
df_sejong_split
```

```python
df_sejong_split.columns = ['시','군','구']
df_sejong_split
```

```python
df_sejong_1 = pd.concat([df_sejong,df_sejong_split], axis = 1)
df_sejong_1
```

## 3. 도담동 => 고운동으로 변경 


```python
df_sejong_1[df_sejong_1['군'] == '도담동']
```

```python
df_sejong_1.loc[df_sejong_1['군']=='도담동','군'] = '고운동'
df_sejong_1[df_sejong_1['군'] == '도담동']
```

## 4. 세종시 오피스텔 거래금액(만원)을 int형으로 변경하세요 


```python
df_sejong_1['거래금액(만원)'] = df_sejong_1['거래금액(만원)'].str.replace(',','')
df_sejong_1['거래금액(만원)'] = df_sejong_1['거래금액(만원)'].astype(int)
df_sejong_1['거래금액(만원)']
```

## 5. 시군구별 전용면적(㎡)별 오피스텔 거래금액 평균 값 구하기 


```python
df_sejong_1.columns[6]
```

```python
df_sejong_1.groupby(df_sejong_1.columns[6]).mean()[['거래금액(만원)']]
```

## 6. 계약년월 별, 평균 오피스텔 가격 구하기  

```python
df_sejong_1
```

```python
df_sejong_1.groupby('계약년월').mean()[['거래금액(만원)']]
```
